# Social Information

```python
# install necessary packages and import them as needed

#!pip install config
#!pip install beautifulsoup4

import pandas as pd
import numpy as np
import requests
import bs4
import urllib3
import config
import time
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

#!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# Remove request warnings
urllib3.disable_warnings()

#define the agent for the web scraping
agent = {"User-Agent":"Mozilla/5.0"}
```

```python
# define the lists to contain the scraped information

names = []
company_names = []
descriptions = []
telephones = []
websites = []
facebooks = []
twitters = []
instagrams = []
weibos = []
googles = []
lines = []
wechats = []
youtubes = []


# define the initial link to get the information needed for Audemars Piguet
AP = 'https://fashionbi.com/brands/audemars-piguet/financials'
# get access using the specified criteria
web = requests.get(AP, verify=True, headers=agent)
time.sleep(8)
# scrape the website using beautiful soup using an html parser
page_soup = soup(web.content, 'html.parser')

# create a wrap for the brand name
h_name = page_soup.find_all('div', {'class': 'mtop10'})
# get the text
name = h_name[0].text
#append to the respective list
names.append(name)

# create a wrap for the company name
c_name = page_soup.find_all('div', {'class': 'col-sm-6'})[1]
# strip the text from unnecessary text
company = [i.strip() for i in c_name if i.name is None and i.strip() != '']
# get the text
company_name  = company[0]
#append to the respective list
company_names.append(company_name)

# create a wrap for the brand description
info = page_soup.find_all('p')
# get the text
description = info[1].text
#append to the respective list
descriptions.append(description)

# create a wrap for the telephone
c_telephone = page_soup.find_all('div', {'class': 'col-sm-6'})[2]
# strip the text from unnecessary text
telephone = [i.strip() for i in c_telephone if i.name is None and i.strip() != '']
# get the text
company_telephone = telephone[0]
#append to the respective list
telephones.append(company_telephone)

# create a wrap for the social information
media = page_soup.find_all('a', {'target':'_blank'})

# get the text
website = media[1].get("href")
#append to the respective list
websites.append(website)
  
# get the text    
facebook = media[2].get("href")
#append to the respective list
facebooks.append(facebook)

# get the text
twitter = media[3].get("href")
#append to the respective list
twitters.append(twitter)

# get the text
instagram = media[4].get("href")
#append to the respective list
instagrams.append(instagram)

# get the text
weibo = media[5].get("href")
#append to the respective list
weibos.append(weibo)

# get the text
google = media[6].get("href")
#append to the respective list
googles.append(google)

# get the text
line = media[7].get("href")
#append to the respective list
lines.append(line)

# get the text
wechat = media[8].get("href")
#append to the respective list
wechats.append(wechat)

# get the text
youtube = media[9].get("href")
#append to the respective list
youtubes.append(youtube)
```
